```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import descartes
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
def fix_crs(map_ob):
    return map_ob.to_crs({'init': 'epsg:4326'})
```

```python
str_map = gpd.read_file("ecological_reserves/BC_Eco_Reserves.shp")
```

```python
str_map = fix_crs(str_map)
```

```python
fig, ax = plt.subplots(figsize = (15,15))
str_map.plot(ax = ax)
```

```python
van = gpd.read_file("Boundaries - Census Divisions 2011/CD_2011.shp")
```

```python
van = fix_crs(van)
```

```python
fig, ax = plt.subplots(figsize = (100,100))
ax.set_ylim(49.002196,49.580115)
ax.set_xlim(-123.490034, -122.387677)
van.plot(ax = ax, color='white', edgecolor='black',)
# van.plot(ax = ax)
# str_map.plot(ax = ax)
```
