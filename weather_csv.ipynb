```python
import pandas as pd
import numpy as np
```

```python
df_meta = pd.read_csv("weather_meta.csv")
```

```python
df_meta = df_meta[['network_name', 'native_id', 'elev', 'lat', 'lon']]
```

```python
df_final = pd.DataFrame(columns=['network_name', 'native_id', 'elev', 'lat', 'lon', 'Tx_Climatology',
                           'Tn_Climatology', 'T_mean_Climatology', 'Precip_Climatology',
                          ])
```

```python
import os
rootdir = 'weather_ascii'
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file[-2:] == "ii":
            curr_path = os.path.join(subdir, file)
            df = pd.read_table(curr_path, skiprows=[0], sep=', ')
            if (df.shape[0] > 0):
                df = df[['Tx_Climatology','Tn_Climatology', 'T_mean_Climatology', 'Precip_Climatology']]
                df = df.replace(["None"], np.NaN)
                df = df.astype(float)

                network_name = subdir.split("/")[1]
                native_id = file.split('.')[0]

                meta_curr = df_meta[(df_meta['network_name'] == network_name) & (df_meta['native_id'] == native_id)].iloc[0]
                val_curr = df.describe().loc["mean"]
                df_temp =  pd.DataFrame( np.hstack((meta_curr.values, val_curr.values)), 
                                        index=np.hstack((meta_curr.index.values, val_curr.index.values))).transpose()

                df_final = df_final.append(df_temp)

            
            
            
```

```python
df_final.shape[0]
```

```python
df_final = df_final[(pd.notnull(df_final['Tx_Climatology'])) |(pd.notnull(df_final['Tn_Climatology'])) |
                   (pd.notnull(df_final['T_mean_Climatology'])) | (pd.notnull(df_final['Precip_Climatology']))]
```

```python
# df_final['elev'] = df_final['elev'].astype(float)
# df_final['lat'] = df_final['lat'].astype(float)
# df_final['lon'] = df_final['lon'].astype(float)

df_final['Tx_Climatology'] = df_final['Tx_Climatology'].astype(float)
df_final['Tn_Climatology'] = df_final['Tn_Climatology'].astype(float)
df_final['T_mean_Climatology'] = df_final['T_mean_Climatology'].astype(float)
df_final['Precip_Climatology'] = df_final['Precip_Climatology'].astype(float)
```

```python
df_final = df_final.reset_index().drop(["index"],1)
```

```python
df_filter_lower = medians - 2*stds
df_filter_upper = medians + 2*stds
```

```python
measures = ['Tx_Climatology','Tn_Climatology', 'T_mean_Climatology', 'Precip_Climatology']
```

```python
indexes_to_drop = set()
for measure in measures:
    df_base = df_final[(pd.notnull(df_final[measure]))]
    df_final_filtered = df_base[~((df_base[measure] > df_filter_lower[measure]) & (df_base[measure] < df_filter_upper[measure]))]
    indexes_to_drop = indexes_to_drop.union(set(df_final_filtered.index))
    
    
```

```python
df_final = df_final.drop(indexes_to_drop)
```

```python
def fix_crs(map_ob):
    return map_ob.to_crs({'init': 'epsg:4326'})
```

```python
sei = gpd.read_file("ShinyMap/MVSEI2014/MVSEI2014.shp")
sei = fix_crs(sei)
```

```python
sei['Tx_Climatology'] = np.nan
sei['Tn_Climatology'] = np.nan
sei['T_mean_Climatology'] = np.nan
sei['Precip_Climatology'] = np.nan
# sei['centroid'] = sei.centroid
```

```python
sei = sei.reset_index().drop(['index'], 1)
weather = weather.reset_index().drop(['index'], 1)
```

```python
df_distances = pd.DataFrame(columns=weather.index)
```

```python
df_distances
```

```python
for i in range(0, weather.shape[0]):
    df_distances[i] = sei.distance(weather.loc[i].geometry)
```

```python
def dictFull(d):
    num = 1
    for col, stat in d.iteritems():
        if (stat == 0):
            num = 0
            break
    return num
```

```python
for i in range(0, sei.shape[0]):
    info_dict = {'Tx_Climatology':0,'Tn_Climatology':0, 'T_mean_Climatology':0, 'Precip_Climatology':0}
    for ind, dist in df_distances.loc[i].sort_values().iteritems():
        if not dictFull(info_dict):
            for col, stat in info_dict.iteritems():
                if not info_dict[col]:
                    if not np.isnan(weather.loc[ind][col]):
                        sei.at[i, col] = weather.loc[ind][col]
                        info_dict[col] = 1
```
